## sigmoid function

```python
#sigmoid function is mainly used in logistic regression 
#it mainly convert any value into probability between 0 to 1

#Suppose we want to predict whether a customer will buy a product:
# 0 = No
# 1 = Yes
# A model may produce a value like: -5, 2.3, 8.1, -1.7
#After applying sigmoid:
# #-5   → 0.0067
#  2.3 → 0.91
#  8.1 → 0.9997
# -1.7 → 0.15
#by using these numbers we can know the probabiliy
#example 0.0067 means 6.6% of chance.
#if probability >= 0.5:
#     predict 1
# else:
#     predict 0
```

```python
#sigmoid function now and do the math with hand
import math
def sigmoid(x):
    return 1/(1+ math.exp(-x))           
```

```python
def prediction_function(age):
    z=0.054* age - 2.13
    y=  sigmoid(z)
    return y
```

```python
age=35
prediction_function(age)
#here 0.44 is less than 0.50 which means low chance of taking insurance.
```

```python
age=47
prediction_function(age)
#here 0.60 is grater than 0.50 which means high chance of taking insurance.
```

## confusion matrix

```python
#confussion matrix
#A Confusion Matrix is a table used to evaluate the performance of a classification model.
# #|                     | Predicted Positive  | Predicted Negative  |
# | ------------------- | ------------------- | ------------------- |
# | **Actual Positive** | True Positive (TP)  | False Negative (FN) |
# | **Actual Negative** | False Positive (FP) | True Negative (TN)  |

#true positive(tp) --------> model predicts "yes" ,acutal answer "yes"
#true negative(tn) --------> model predicts "no",acutal answer "no"
#false negative(fn) -------> model predicts "yes", acutal answer "no"
#false postive(fp) --------> model predicts "no",actual answer "yes".
```

```python
y_predicted=model.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,y_predicted)    #here x value test ullla 360 data use cheyth predicted data vech use cheyaa.
cm
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm= confusion_matrix(y_test,y_predicted)

disp=ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="nipy_spectral")

plt.show()
```
